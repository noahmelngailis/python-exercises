```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = range(-10,11)
y = [(x**2 - x + 2) for x in x]

plt.plot(x,y)
plt.annotate('Origin', xy=(0, 0), xytext=(-3, 5), arrowprops={'facecolor': 'red'})
```

```python
import math as m
x1 = range(1,10000)
y1 = [m.sqrt(x) for x in x1]

plt.plot(x1, y1)
plt.title("Graph of square root function")
```

```python
x2 = range(-10,11)
y2 = [x**3 for x in x2]

plt.plot(x2, y2)
plt.title("Graph of cubic function")
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
```

```python
x3 = range(-100,101)
y3 = [m.tan(x) for x in x3]

plt.plot(x3,y3)
plt.title("Graph of tan function")
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
```

```python
x4 = range(-10,11)
y4 = [2**x for x in x4]

plt.plot(x4, y4)
plt.title("Graph of 2^x function")
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
```

```python
import math as m
x1 = range(1,11)
y1 = [m.sqrt(x) for x in x1]

x2 = range(-10,11)
y2 = [x**3 for x in x2]

x3 = range(-10,11)
y3 = [m.tan(x) for x in x3]

x4 = range(-10,11)
y4 = [2**x for x in x4]

plt.subplot(2, 2, 1)
plt.plot(x1, y1)

plt.subplot(2, 2, 2)
plt.plot(x2, y2)

plt.subplot(2, 2, 3)
plt.plot(x3, y3)

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.suptitle("Four functions")
```

```python
import math as m
x1 = range(1,11)
y1 = [m.sqrt(x) for x in x1]

x2 = range(-10,11)
y2 = [x**3 for x in x2]

x3 = range(-10,11)
y3 = [m.tan(x) for x in x3]

x4 = range(-10,11)
y4 = [2**x for x in x4]

plt.plot(x1,y1, label = 'sqrt(x)')
plt.plot(x2, y2, label = 'x**3')
plt.plot(x3, y3, label = 'tan(x)')
plt.plot(x4, y4, label = '2**x')
plt.title("A useless graph")
plt.legend()
```
