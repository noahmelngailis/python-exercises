```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = range(-10,11)
y = [(x**2 - x + 2) for x in x]

plt.plot(x,y)
plt.annotate('Origin', xy=(0, 0), xytext=(-3, 5), arrowprops={'facecolor': 'red'})
```

```python
import math as m
x1 = range(1,10000)
y1 = [m.sqrt(x) for x in x1]

plt.plot(x1, y1)
plt.title("Graph of square root function")
```
